```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
```

```python
df = pd.read_csv("winequality-red.csv")
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
df = df[~(df['residual sugar']>(df['residual sugar'].mean()+df['residual sugar'].std()*2))]
```

```python
X = df.drop(['quality'],axis=1) 
y = df['quality']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler

# Initialize the StandardScaler
scaler = StandardScaler()

# Fit the StandardScaler to the training data
scaler.fit(X_train)

# Use the scaler to transform the training and test sets
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
model = GradientBoostingRegressor(random_state=42)
model.fit(X_train_scaled,y_train)
```

```python
y_pred =model.predict(X_test_scaled)
```

```python

print(mean_absolute_error(y_test,y_pred))
print(np.sqrt(mean_squared_error(y_test,y_pred)))
```

```python
from sklearn.model_selection import GridSearchCV
params = {

    'loss' : ['squared_error', 'absolute_error'],
    'n_estimators' : [10,50,100],
    'criterion' : ['squared_error'],
    'ccp_alpha' : [0,1,5,10]

} 
gsv = GridSearchCV(model,param_grid=params,scoring='accuracy',cv=5)
```

```python
gsv.fit(X_train_scaled,y_train)
```

```python
gsv.best_params_
```

```python
model2 = GradientBoostingRegressor(ccp_alpha= 1,criterion = 'squared_error',loss ='absolute_error',n_estimators=10)
```

```python
model2.fit(X_train_scaled,y_train)
```

```python
y_pred2 = model2.predict(X_test_scaled)
```

```python
print('MAE = ',mean_absolute_error(y_test,y_pred2))
print('MSE = ',np.sqrt(mean_squared_error(y_test,y_pred2)))
```
